### Q01.5 寫一函式去除所有中文停用詞`removeChiStopWords()`
```
def removeChiStopWords(...):
    ...

filtered = removeChiStopWords(filtered)
print(filtered)
```

````python
stopwords = HanziConv.toSimplified(stopwords)
# print(stopwords[:100])
key_to_remove = []
def removeChiStopWords(word_dict):
  for k, v in word_dict.items():
    if k in stopwords:
      key_to_remove.append(k)
  
  for i in range(len(key_to_remove)):
    del word_dict[key_to_remove[i]]

  return word_dict

filtered = removeChiStopWords(filtered)
# print(key_to_remove)
print(filtered)
# print(len(filtered))
````

### Q01.6 寫一函式去除所有的英文字母與數字`removeAlphaNum()`。
```
def removeAlphaNum(...):
    ...
filtered = removeAlphaNum(filtered)
print(filtered)
```

````python
# YOUR CODE SHOULD BE HERE
# YOU CAN CHANGE THE VARIABLE NAME
import string
key_to_remove2 = []

punc = "0123456789"+string.ascii_lowercase[:]+string.ascii_uppercase[:]

def removeAlphaNum(word_dict):
  for k, v in word_dict.items():
    for i in range(len(k)):
      if k[i] in punc:
        key_to_remove2.append(k)
        break
        
  # print(key_to_remove2)
  for i in range(len(key_to_remove2)):
    del word_dict[key_to_remove2[i]]

  return word_dict


filtered = removeAlphaNum(filtered)
print(filtered)
# print(len(filtered))
````

### Q01.7 寫一函式去除全半形標點符號或空白：`removalPuncSpace()`。
```
def removalPuncSpace(...):
    ...
filtered = removalPuncSpaces(filtered)
print(filtered)
```

````python
key_to_remove3 = []
stops = stops+"⋯"+"\n"

def removalPuncSpace(word_dict):
  for k, v in word_dict.items():
    for i in range(len(k)):
      if k[i] in punctuation:
        key_to_remove3.append(k)
        break
      if k[i] in non_stops:
        key_to_remove3.append(k)
        break
      if k[i] in stops:
        key_to_remove3.append(k)
        break
        
  # print(key_to_remove3)
  for i in range(len(key_to_remove3)):
    del word_dict[key_to_remove3[i]]
  return word_dict  

filtered = removalPuncSpace(filtered)
print(filtered)
# print(len(filtered))

# re.sub("[\s+\.\!\/_,$%^*(+\"\']+|[+——！，。？、~@#￥%……&*（）]", "",corpus)
````
